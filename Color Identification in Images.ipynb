### Project: Color Identification in Images

```python
# importing libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading the image using opencv

img = cv2.imread(r"C:\Users\varun\OneDrive\Desktop\Object Detection\colours.JPG")
```

```python
# display the image using matplotlib

plt.figure(figsize=(20,8))
plt.imshow(img)
```

### The above image doesnot look with the true colors it has too. It is because the image was displayed in BGR

```python
# convert the image BGR into RGB

grid_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# plot the image agian

plt.figure(figsize=(20,8))
plt.imshow(grid_RGB)
```

```python
# opencv  in order to detect the image it uses HSV color format instead of RGB or BGR

grid_HSV = cv2.cvtColor(grid_RGB, cv2.COLOR_RGB2HSV)
```

```python
# HSV range of the color

lower_yellow = np.array([25,150,50])
upper_yellow = np.array([35,255,255])
```

```python
# to show the required color and blackout the rest of the colors

mask = cv2.inRange(grid_HSV, lower_yellow, upper_yellow)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

### The above image is not the orginal yellow color but its the mask

```python
# for displaying the orginal color by removing the mask(use bitwise 'and' operator)

res = cv2.bitwise_and(grid_RGB, grid_RGB, mask=mask)
```

```python
# plot the image

plt.figure(figsize=(20,8))
plt.imshow(res)
```

# Checking the model with other HSV Range color

```python
lower_white = np.array([0,0,255])
upper_white = np.array([0,0,255])
```

```python
mask = cv2.inRange(grid_HSV, lower_white, upper_white)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(res)
```
